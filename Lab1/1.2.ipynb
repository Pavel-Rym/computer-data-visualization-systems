```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Завантаження даних та виведення перших рядків

```python
df = pd.read_csv('shopping_trends.csv')
df.head()
```

Перевірка розміру таблиці, наявність пропусків, дублікатів, тип стовпців

```python
print(df.info())
```

Перейменування стовпців

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

Гістограма віку (age)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Порівняння розподілу покупок за статтею (item_purchased)

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='item_purchased', data=df, order=df['item_purchased'].value_counts().index)
plt.title('Items Purchased Distribution')
plt.xlabel('Count')
plt.ylabel('Item Purchased')
plt.show()
```

Залежність між віком та сумою покупок (purchase_amount)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='purchase_amount_(usd)', data=df)
plt.title('Age vs Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

Сезонність покупок (season)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='season', data=df)
plt.title('Seasonal Purchases')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

Графік розподілу розмірів одягу (size)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='size', data=df)
plt.title('Size Distribution')
plt.xlabel('Size')
plt.ylabel('Count')
plt.show()
```

Кількість покупок відносно типу оплати (payment_method)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='payment_method', data=df)
plt.title('Payment Method Distribution')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

Залежність між кількістю попередніх покупок (previous_purchases) і сумою поточної покупки (purchase_amount)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='previous_purchases', y='purchase_amount_(usd)', data=df)
plt.title('Previous Purchases vs Purchase Amount')
plt.xlabel('Previous Purchases')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Створення діапазонів для x та y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Створення координатної сітки
X, Y = np.meshgrid(x, y)

# Обчислення значень для z
Z = np.sqrt(np.sin(X**2 + Y**2))

# Побудова 3D графіку
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Додавання назви графіку та підписів осей
ax.set_title('3D Surface Plot of sqrt(sin(x^2 + y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Додавання кольорової легенди
fig.colorbar(surface, shrink=0.5, aspect=5)

# Відображення графіку
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Генерація даних у вигляді матриці (наприклад, кореляційної матриці)
data = np.random.rand(10, 10)

# Побудова графіка теплової поверхні (heatmap)
plt.figure(figsize=(10, 6))
sns.heatmap(data, annot=True, cmap='viridis')

# Додавання назви графіку та підписів осей
plt.title('Heatmap of Random Data')
plt.xlabel('X')
plt.ylabel('Y')

# Відображення графіку
plt.show()
```
